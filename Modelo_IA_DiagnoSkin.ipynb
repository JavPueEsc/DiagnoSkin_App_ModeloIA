```python
# Importaciones necesarias para usar Keras con TensorFlow 2.x
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Aqu칤 van el resto de las importaciones necesarias para el proyecto
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2

# Cargar MobileNetV2 como modelo base (sin usar TensorFlow Hub)
mobilenetv2_base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Congelar las primeras capas del modelo base
mobilenetv2_base.trainable = False
```

```python
# Crear un modelo de la red con MobileNetV2 como base
modelo = Sequential([
    mobilenetv2_base,
    GlobalAveragePooling2D(),
    Dense(5, activation='softmax')  # Nueva capa de salida
])
```

```python
# Resumen del modelo
modelo.summary()
```

```python
# Compilar el modelo utilizando la API de optimizadores de TensorFlow 2.x
modelo.compile(
    optimizer=Adam(learning_rate=0.0001),  # Ajuste de la tasa de aprendizaje para fine-tuning
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Crear generadores de datos para entrenamiento y validaci칩n
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 20% de las im치genes ser치n de validaci칩n
)

# Directorio con las im치genes
ruta_datos = "datasetPruebaMEL_NV"

# Generador para entrenamiento
data_gen_entrenamiento = datagen.flow_from_directory(
    ruta_datos,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# Generador para validaci칩n
data_gen_pruebas = datagen.flow_from_directory(
    ruta_datos,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

# Fine-Tuning: Desbloquear las 칰ltimas capas de MobileNetV2
for layer in mobilenetv2_base.layers[-10:]:  # Ajustar el n칰mero de capas seg칰n el modelo
    layer.trainable = True

# Volver a compilar el modelo despu칠s de habilitar el fine-tuning
modelo.compile(
    optimizer=Adam(learning_rate=0.00007),  # Ajustar la tasa de aprendizaje para fine-tuning
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callback para guardar solo el mejor modelo seg칰n val_accuracy
checkpoint = ModelCheckpoint(
    filepath='modelo_mejor.keras',
    monitor='val_accuracy',
    mode='max',
    save_best_only=True,
    verbose=1
)

# Entrenar con callback para guardar solo mejor modelo
EPOCAS = 75
historial = modelo.fit(
    data_gen_entrenamiento,
    epochs=EPOCAS,
    validation_data=data_gen_pruebas,
    callbacks=[checkpoint]
)
'''
# Entrenar el modelo
EPOCAS = 75
historial = modelo.fit(
    data_gen_entrenamiento, epochs=EPOCAS,
    validation_data=data_gen_pruebas
)
# Guardar el modelo entrenado
 modelo.save("modelo.keras")  # Aqu칤 se guarda el modelo
'''
```

```python
import requests
import numpy as np
import cv2
from tensorflow.keras.models import load_model

# Carga del modelo .keras
modelo = load_model("modelo_mejor.keras") 

# Lista de clases, en el orden en que el modelo las espera
clases = ['MEL', 'NV', 'VAS', 'XDF','ZNF']

def categorizar(url):
    try:
        # Descargar la imagen
        respuesta = requests.get(url)
        respuesta.raise_for_status()  # lanza error si falla la descarga

        # Convertir a imagen
        img_array = np.array(bytearray(respuesta.content), dtype=np.uint8)
        img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

        # Preprocesar imagen
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (224, 224))
        img = img.astype("float32") / 255.0
        img = np.expand_dims(img, axis=0)  # A침adir dimensi칩n batch

        # Hacer la predicci칩n
        prediccion = modelo.predict(img)
        probabilidad_max = np.max(prediccion)
        indice_prediccion = np.argmax(prediccion)

        clase_predicha = clases[indice_prediccion]
        porcentaje_confianza = round(probabilidad_max * 100, 2)

        return f"Clase predicha: {clase_predicha} ({porcentaje_confianza}%)"

    except Exception as e:
        return f"Error al procesar la imagen: {e}"
```

# Hacer predicciones de im치genes del dispositivo

```python
from tensorflow.keras.preprocessing import image
import os

# Cargar el modelo
modelo = tf.keras.models.load_model("modelo_mejor.keras")  # o "modelo.h5" o carpeta

# Tama침o esperado por el modelo
IMG_SIZE = (224, 224)  # Ajusta seg칰n lo que tu modelo espere

# Diccionario opcional de clases, si lo tienes
CLASES = ['MEL', 'NV', 'VAS', 'XDF','ZNF']  # Personaliza seg칰n tu dataset

def predecir_imagen_ruta(ruta_imagen):
    if not os.path.exists(ruta_imagen):
        print(f"No se encontr칩 la imagen en la ruta: {ruta_imagen}")
        return

    # Cargar la imagen
    img = image.load_img(ruta_imagen, target_size=IMG_SIZE)
    
    # Convertir a array y escalar
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0  # Normalizaci칩n si tu modelo lo requiere
    img_array = np.expand_dims(img_array, axis=0)  # A침adir batch dimension

    # Hacer la predicci칩n
    predicciones = modelo.predict(img_array)

    # Si es clasificaci칩n binaria o multiclase
    clase_predicha = np.argmax(predicciones[0])
    confianza = np.max(predicciones[0]) * 100

    # Mostrar resultado
    print(f"Predicci칩n: {CLASES[clase_predicha]} ({confianza:.2f}%)")

    return CLASES[clase_predicha], confianza
```

# Convertir a tensorflow lite optimizado para m칩viles

```python
import tensorflow as tf

# 1. Carga tu modelo Keras (.h5 o .keras)
modelo = tf.keras.models.load_model("modelo_mejor.keras")  # o "modelo.h5"

# 2. Crea el convertidor
converter = tf.lite.TFLiteConverter.from_keras_model(modelo)

# 3. Activa optimizaciones (reduce tama침o y mejora rendimiento)
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# 4. Especifica tipo de cuantizaci칩n: float16 (m치s compatible y r치pido)
converter.target_spec.supported_types = [tf.float16]

# 5. Convierte a TFLite
tflite_model = converter.convert()

# 6. Guarda el modelo optimizado
with open("modelo_mejor_optimizado.tflite", "wb") as f:
    f.write(tflite_model)

print("九 Conversi칩n completada y modelo optimizado guardado como modelo_mejor_optimizado.tflite")
```
